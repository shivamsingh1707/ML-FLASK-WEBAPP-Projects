```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data.head(2)
```

```python
data['Patient Id'] = le.fit_transform(data['Patient Id'])
data['Level'] = le.fit_transform(data['Level'])
data.head()
```

```python
y = data['Level']
x= data.drop(columns = ['Level'])
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
names = x.columns

```

```python
x = scale.fit_transform(x)
x
```

```python
x = pd.DataFrame(x,columns = names)
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
logi = LogisticRegression()
logi.fit(x_train,y_train)
```

```python
ytr = logi.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

print(accuracy_score(y_train,ytr))
```

```python
ytt = logi.predict(x_test)
print(accuracy_score(y_test,ytt))
```

```python
#RandomForest
from sklearn.ensemble import RandomForestClassifier
rand = RandomForestClassifier()
rand.fit(x_train,y_train)
```

```python
ytr = rand.predict(x_train)
print(accuracy_score(y_train,ytr))
```

```python
ytt = rand.predict(x_test)
print(confusion_matrix(y_train,ytr))
```

```python
print(confusion_matrix(y_test,ytt))
```

```python
print(accuracy_score(y_test,ytt))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
```

```python
#Decision tree
dtree.fit(x_train,y_train)
ytr = dtree.predict(x_train)
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(y_train,ytr))
```

```python
print(confusion_matrix(y_train,ytr))
```

```python
ytt = dtree.predict(x_test)
print(accuracy_score(y_test,ytt))
```

```python
print(confusion_matrix(y_test,ytt))
```
